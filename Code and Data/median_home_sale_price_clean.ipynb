```python
import pandas as pd
```

```python
raw = pd.read_csv('Metro_median_sale_price_uc_sfrcondo_sm_sa_month.csv')
raw.head()
```

```python
# Make USA region (row 1) be transferred into a "state"
# Change StateName value in Row 0 to be USA, then drop RegionName column
raw.at[0, 'StateName'] = 'USA'
# print(raw.iloc[0]['StateName'])
raw = raw.drop(['SizeRank', 'RegionType', 'RegionID', 'RegionName'], axis=1)

raw.head()
```

```python
# Melt raw on StateName
raw = pd.melt(raw, id_vars=['StateName'])
# raw.variable.unique()
raw.head()
```

```python
# Rename columns to ['state_name', 'date', 'median_home_price']
raw.rename(columns={'StateName': 'state_name', 'variable': 'date', 'value': 'median_home_price'}, inplace=True)
raw.head()
```

```python
# Group raw by state_name and date
# Take median of aggregation on median_home_price
clean = raw.groupby(['state_name', 'date'])
clean = clean['median_home_price'].median()
clean = clean.reset_index()

# clean.head()
```
